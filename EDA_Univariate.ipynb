## c. Boxplot

```python
sns.boxplot(x='median_house_value',data=df)
plt.xlabel('Median House Value')
plt.title('Distribution of Median House Value')
```

```python
sns.boxplot(df['median_income'])
plt.title('Box Plot of Median Income')
plt.ylabel('Median Income')
plt.show()
```

This code generates a box plot for the 'median_income' column in your DataFrame. A box plot is useful for visualizing the distribution of numerical data and detecting outliers. The box represents the interquartile range (IQR), the line inside the box is the median, and the "whiskers" extend to 1.5 times the IQR from the quartiles. Points outside the whiskers are considered outliers.
